```python
# removing samples since sample cost $0.01 
filtered_df = df_required_columns[df_required_columns['Listing Price'] != '$0.01']
```

```python
filtered_df
```

```python
filtered_df.to_csv('filetered.csv',index = False)
```

```python
#converting units into packages
# 1package is 35 units 
```

```python
filtered_df.loc[filtered_df['Weight/Unit'] == 'Units', 'Quantity'] = filtered_df['Quantity'] / 35
filtered_df.loc[filtered_df['Weight/Unit'] == 'Units', 'Weight/Unit'] = 'Packages'

# Round package quantities and change dtype to int if no fractional packages
filtered_df['Quantity'] = filtered_df['Quantity'].apply(lambda x: int(x) if x.is_integer() else x)
filtered_df['Quantity'] = filtered_df['Quantity'].round(2)
```

```python
filtered_df['Order Date'] = pd.to_datetime(filtered_df['Order Date'])
```

```python
filtered_df['Year-Week'] = filtered_df['Order Date'].dt.strftime('%Y-%U')
```

```python
# i had to discard invoice number Good_F-158, since i had no idea about the cost of the goods here.
filtered_df = filtered_df[filtered_df['Invoice Number'] != 'Good_F-158']
```

```python
filtered_df.to_csv('new_one_with_date_modified.csv',index = False)
```

```python
weekly_sales = filtered_df.groupby(['Product Name', 'Buyer Address', 'Year-Week'])['Quantity'].sum().reset_index()
```

```python
weekly_sales
```

```python
weekly_sales.to_csv('weekly_store.csv')
```

```python
total_sales_per_product_store = weekly_sales.groupby(['Product Name', 'Buyer Address'])['Quantity'].sum().reset_index()
```

```python
total_sales_per_product_store
```
